description:
>The dataset represents 10 years (1999-2008) of clinical care at 130 US hospitals and integrated delivery networks. It includes over 50 features representing patient and hospital outcomes. Information was extracted from the database for encounters that satisfied the following criteria.
>>(1) It is an inpatient encounter (a hospital admission).<br>
(2)	It is a diabetic encounter, that is, one during which any kind of diabetes was entered to the system as a diagnosis.<br>
(3)	The length of stay was at least 1 day and at most 14 days.<br>
(4)	Laboratory tests were performed during the encounter.<br>
(5)	Medications were administered during the encounter.

The data contains such attributes as patient number, race, gender, age, admission type, time in hospital, medical specialty of admitting physician, number of lab test performed, HbA1c test result, diagnosis, number of medication, diabetic medications, number of outpatient, inpatient, and emergency visits in the year before the hospitalization, etc.

Feature Engineering:<br>
Dropping redundant/sparse columns<br>
Filling columns<br>
Converting categorical/nominal features to numeric<br>

Outputs from here:<br>
Features [dataframe]<br>
outputs [dataframe]

```python
# Drop columns of sparse data
# weight, payer_code, medical_specialty
new_df = df.drop(columns=['weight','payer_code','medical_specialty'])

# Drop redundant columns
# encounter_id, patient_nbr
new_df = new_df.drop(columns=['encounter_id','patient_nbr'])

# Drop columns with too many categories (to prevent memory error)
# diag_1, diag_2, diag_3
new_df = new_df.drop(columns=['diag_1','diag_2','diag_3'])

# Find more columns with missing data
# Missing data is represented as '?' instead of null values in this dataset
new_df.isin(['?']).sum()
# Columns with missing values:
# race, diag_1, diag_2, diag_3
# (all of the features with missing data are nominal features and are already filled with a speicific string, '?', so there are no numeric features that need conversion nor columns that need tobe filled)
# Fill missing columns (unrequired)

# Split features and targets
outputs = new_df[['readmitted']].copy() # target
new_df = new_df.drop(columns=['readmitted']) # features

# Split features (numeric and categorical)
# numeric: time_in_hospital, num_lab_procedures, num_procedures, num_medications, number_outpatient, number_emergency, number_inpatient, number_diagnoses
# categorical: the rest
new_df_num = new_df[['time_in_hospital','num_lab_procedures','num_medications','number_outpatient','number_emergency','number_inpatient','number_diagnoses']].copy()
new_df_cat = new_df.drop(columns=['time_in_hospital','num_lab_procedures','num_medications','number_outpatient','number_emergency','number_inpatient','number_diagnoses'])

# convert categorical features to numeric
enc = OneHotEncoder(sparse=False)
new_df_cat = pd.DataFrame(enc.fit_transform(new_df_cat))

# merge numeric and categorical dataframes
Features = pd.concat([new_df_num, new_df_cat], axis=1, sort=False)

# we want to predict readmission within 30 days as according to the academic article linked above so group '>30' and 'NO' together
outputs = outputs.replace(to_replace='>30', value='NO')
# Now the outputs are either 'NO' or <30
```

Pre-train Conditions:<br>
Train test split<br>
Feature selection<br>
Model space<br>
Hyperparameter space

```python
# training set, testing set split
X_train, X_test, y_train, y_test = train_test_split(Features, outputs, test_size=0.2, random_state=random_seed)

"""# feature Selection - model based feature selection via recursion and cross validation
skf = StratifiedKFold(n_splits=5, random_state=random_seed)
selection = RFECV(RandomForestClassifier(random_state=random_seed), cv=skf, scoring=score)
selection.fit(X_train, y_train)
X_train_selection = selection.transform(X_train)
X_test_selection = selection.transform(X_test)"""
# feature selection is taking too long, just use all features
X_train_selection = X_train
X_test_selection = X_test

skf = StratifiedKFold(n_splits=5, random_state=random_seed)

# Model space    
models = {
    "DecisionTree": DecisionTreeClassifier(random_state=random_seed),
    "NeuralNetwork": MLPClassifier(solver='sgd', learning_rate='adaptive', max_iter=10, random_state=random_seed),
    "NearestNeighbor": KNeighborsClassifier(weights='distance'),
    "Perceptron": Perceptron(random_state=random_seed)
}

# Hyperparamater space
hyperparameters = {
    "DecisionTree":[{'max_depth':[3,4,5,6]}],
    "NeuralNetwork":[{'hidden_layer_sizes':[(5,),(10,)]}],
    "NearestNeighbor": [{'n_neighbors':[5,7,9,11]}],
    "Perceptron": [{'alpha':[0.001,0.01]}]
}

""" The model and hyperparameter spaces have been changed to reduce computing load so the results_report detailed in the next text block are not associated with these models """
```

Training and Evaluating:<br>
Model fitting<br>
Evaluation<br>
Report

```python
# define report
results_report = []

# Model fitting
for identifier, model in models.items():
    
    print(identifier)
    
    # run through different combinations of hyperparameters (with stratified K folds) to identify the best
    classifier = GridSearchCV(model, hyperparameters[identifier], cv=skf, scoring=score)
    classifier.fit(X_train_selection, y_train)
    
    # get the classifier with the best combinations of hyperparameters
    best_model = classifier.best_estimator_
    
    # validate the accuracy
    y_pred = best_model.predict(X_test_selection)
    acc = accuracy_score(y_test, y_pred)
    
    # evaluate accuracy of model
    evaluation = {'model': identifier, 'parameters': str(best_model.get_params()), 'accuracy': acc}
    
    # append to report
    results_report.append(pd.DataFrame(evaluation,index=[0]))
    
# output report
results_report
```
